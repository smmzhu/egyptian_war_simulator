```python
import seaborn as sns
import pandas as pd
import random
import numpy as np
p1_win_rate = 1 #CHANGE: this is rate that p1 wins, 0 means no wins, 1 means 100% win rate, 0.5 = 50%, etc
def check_slap():
    global stack
    global p1_win_rate
    global winner
    global slap_counter
    slapCase = False
    if len(stack) >= 2:
        if stack[-1] == stack[-2]:
            slapCase = True
    if len(stack) >= 3:
        if stack[-1] == stack[-3]:
            slapCase = True
    if slapCase == True:
        if random.random() > p1_win_rate:
            winner = "p1"
        else:
            winner = "p2"
        slap_counter += 1
    return slapCase

all_cards = []
all_slaps = []
p1_wins = []
for foo in range(10000): #looper for 10000 games
    deck = []
    for num in range(2,11):
        for foo in range(4):
            deck.append(str(num))
    for card in ["J", "Q", "K", "A"]:
        for fuck in range(4):
            deck.append(card)
    random.shuffle(deck)
    p1_deck = deck[:26]
    p2_deck = deck[26:]
    winner = None
    turn = "p1"
    cards_played = 0
    slap_counter = 0
    gameRunning = True
    while len(p1_deck) > 0 and len(p2_deck) > 0 and gameRunning: #one game
        stack = []
        if winner == "p1":
            x = p2_deck.copy()
            p2_deck = p1_deck.copy()
            p1_deck = x.copy()
        while True: #one stack
            winner = None
            if turn == "p1":
                if len(p1_deck) > 0 and len(p2_deck) > 0:
                    stack.append(p1_deck.pop(0))
                    cards_played += 1
                else:
                    gameRunning = False
                    break
            elif turn == "p2":
                if len(p1_deck) > 0 and len(p2_deck) > 0:
                    stack.append(p2_deck.pop(0))
                    cards_played += 1
                else:
                    gameRunning = False
                    break
                
            check_slap()
            if stack[-1] in ["J","Q","K","A"]: 
                face_card = stack[-1]
                if turn == "p1":
                    switch = 1
                elif turn == "p2":
                    switch = 2
                running = True
                while running:
                    if switch == 1:
                        if face_card == "J":
                            for foo in range(1):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p2_deck.pop(0))
                                    cards_played += 1
                                else:
                                    gameRunning = False
                                    running = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 2
                                    break
                            if switch == 1 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p1"
                                running = False
                        elif face_card == "Q":
                            for foo in range(2):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p2_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 2
                                    break
                            if switch == 1 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p1"
                                running = False
                        elif face_card == "K":
                            for foo in range(3):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p2_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 2
                                    break
                            if switch == 1 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p1"
                                running = False
                        elif face_card == "A":
                            for foo in range(4):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p2_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break       
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 2
                                    break
                            if switch == 1 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p1"
                                running = False
                    elif switch == 2:
                        if face_card == "J":
                            for foo in range(1):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p1_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 1
                                    break
                            if switch == 2 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p2"
                                running = False
                        elif face_card == "Q":
                            for foo in range(2):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p1_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 1
                                    break
                            if switch == 2 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p2"
                                running = False
                        elif face_card == "K":
                            for foo in range(3):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p1_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 1
                                    break
                            if switch == 2 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p2"
                                running = False
                        elif face_card == "A":
                            for foo in range(4):
                                if len(p1_deck) > 0 and len(p2_deck) > 0:
                                    stack.append(p1_deck.pop(0))
                                    cards_played += 1
                                else:
                                    running = False
                                    gameRunning = False
                                    break
                                if check_slap():
                                    running = False
                                    break
                                if stack[-1] in ["J","Q","K","A"]:
                                    face_card = stack[-1]
                                    switch = 1
                                    break 
                            if switch == 2 and winner == None and len(p1_deck) > 0 and len(p2_deck) > 0:
                                winner = "p2"
                                running = False
        #code smth that says if winner none then flip the winner
            if winner == None:
                if turn == "p1":
                    turn = "p2"
                else:
                    turn = "p1"
            else:   
                if winner == "p1":
                    p1_deck += stack
                    stack = []
                    turn = "p1"
                else:
                    p2_deck += stack
                    stack = []
                    turn = "p2"
    if len(p1_deck) == 0:
        p1_wins.append(1) #1 is a p1 win
    else:
        p1_wins.append(0) #0 is if otherwise
    all_cards.append(cards_played)
    all_slaps.append(slap_counter)




```

```python
data = np.array([all_cards,all_slaps,p1_wins,[1 for x in range(10000)]]) #CHANGE 0.5 or 0.7 or 1 according to win rate
data = data.transpose()
df = pd.DataFrame(data = data, columns = ["cards played", "slaps", "p1_win?","win_rate"] )
sns.histplot(data = df["cards played"], bins = 100, kde = True)
print(df["cards played"].mean())

```
